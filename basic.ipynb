```python
# Examples of generating variables and datatypes in Python

# Integer
age = 25
print(f"Age: {age}, Type: {type(age)}")

# Float
height = 5.9
print(f"Height: {height}, Type: {type(height)}")

# String
name = "Alice"
print(f"Name: {name}, Type: {type(name)}")

# Boolean
is_student = True
print(f"Is student: {is_student}, Type: {type(is_student)}")

# List
fruits = ["apple", "banana", "cherry"]
print(f"Fruits: {fruits}, Type: {type(fruits)}")

# Dictionary
person = {"name": "Bob", "age": 30}
print(f"Person: {person}, Type: {type(person)}")

# Tuple
coordinates = (10, 20)
print(f"Coordinates: {coordinates}, Type: {type(coordinates)}")

# Set
unique_numbers = {1, 2, 3, 4}
print(f"Unique numbers: {unique_numbers}, Type: {type(unique_numbers)}")
```

```python
# Example of control flow using if statement
if age > 20:
    print(f"{name} is older than 20.")
else:
    print(f"{name} is 20 or younger.")
```

```python
# Example of if-else ladder using multiple conditions
if age < 18:
    print(f"{name} is a minor.")
elif age < 30:
    print(f"{name} is a young adult.")
elif age < 65:
    print(f"{name} is an adult.")
else:
    print(f"{name} is a senior.")

# Another if-else ladder example
if is_student:
    print(f"{name} is currently a student.")
    if age < 25:
        print("and is under 25 years old.")
    else:
        print("and is 25 or older.")
else:
    print(f"{name} is not a student.")
```

```python
import matplotlib.pyplot as plt

# Create a simple bar plot using the fruits list
plt.figure(figsize=(8, 5))
plt.bar(fruits, [3, 2, 4])
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.title('Fruit Quantities')
plt.show()
```

```python
# Example of control flow using if-elif-else statement
if age < 18:
    print(f"{name} is a minor.")
elif age < 65:
    print(f"{name} is an adult.")
else:
    print(f"{name} is a senior.")
```

```python
# Example of while loop
count = 0
while count < len(fruits):
    print(f"Fruit {count + 1}: {fruits[count]}")
    count += 1

# Another while loop example with age
remaining_years = 65 - age
year = 0
while year < remaining_years:
    print(f"In {year} years, {name} will be {age + year} years old.")
    year += 5
```

```python
# Example of for loop iterating through the fruits list
for fruit in fruits:
    print(f"Fruit: {fruit}")

# For loop with index using enumerate
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")

# For loop iterating through dictionary items
for key, value in person.items():
    print(f"{key}: {value}")

# For loop iterating through a set
for number in unique_numbers:
    print(f"Number: {number}")
```

```python
# Example of do-while loop simulation in Python
i = 0
while True:
    print(f"Iteration {i + 1}: {fruits[i]}")
    i += 1
    if i >= len(fruits):
        break
```
