```python
import numpy as np

import matplotlib.pyplot as plt

x = np.arange(1, 11)
y1 = np.sin(x / 2)
y2 = np.cos(x / 3)
values = [5, 7, 3, 8, 6]
categories = ["A", "B", "C", "D", "E"]

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x, y1, marker="o", label="sin(x/2)")
axs[0, 0].plot(x, y2, marker="s", label="cos(x/3)")
axs[0, 0].set_title("Line chart")
axs[0, 0].legend()
axs[0, 0].grid(True)

axs[0, 1].bar(categories, values, color="skyblue")
axs[0, 1].set_title("Bar chart")
axs[0, 1].set_ylabel("Value")

axs[1, 0].scatter(x, y1, c=y2, cmap="viridis", s=80)
axs[1, 0].set_title("Scatter plot")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("sin(x/2)")

axs[1, 1].plot(x, np.cumsum(values[:10]), linestyle="--", color="green")
axs[1, 1].set_title("Cumulative sum demo")
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
values2 = [4, 9, 2, 7, 5]
x2 = np.arange(1, 11)
y1_new = np.sin(x2 / 1.8)
y2_new = np.cos(x2 / 2.2)

fig2, axs2 = plt.subplots(2, 2, figsize=(10, 8))
axs2[0, 0].plot(x2, y1_new, marker="o", label="sin(x/1.8)")
axs2[0, 0].plot(x2, y2_new, marker="s", label="cos(x/2.2)")
axs2[0, 0].set_title("Line chart")
axs2[0, 0].legend()
axs2[0, 0].grid(True)

axs2[0, 1].bar(categories, values2, color="salmon")
axs2[0, 1].set_title("Bar chart")
axs2[0, 1].set_ylabel("Value")

axs2[1, 0].scatter(x2, y1_new, c=y2_new, cmap="plasma", s=80)
axs2[1, 0].set_title("Scatter plot")
axs2[1, 0].set_xlabel("x")
axs2[1, 0].set_ylabel("sin(x/1.8)")

axs2[1, 1].plot(np.arange(1, len(values2) + 1), np.cumsum(values2), linestyle="--", color="green")
axs2[1, 1].set_title("Cumulative sum demo")
axs2[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
